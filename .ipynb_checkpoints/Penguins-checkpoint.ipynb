### Search for null values and duplicates

```python
df.isnull().sum()
```

```python
# Replace the null values by the median or mode in the case of qualitative values depending on each species.

numeric_cols = ['bill_length_mm','bill_depth_mm','flipper_length_mm','body_mass_g']
categorical_cols = ['sex']

for species in df['species'].unique():
    # Filter by species.
    species_df = df[df['species'] == species]
    
    # Replace null values in numeric columns with the median.
    for col in numeric_cols:
        species_df[col].fillna(species_df[col].median(), inplace=True)
    
    # Replace null values in categorical columns with mode.
    for col in categorical_cols:
        species_df[col].fillna(species_df[col].mode().iloc[0], inplace=True)
        
    # Update the original dataframe with the replaced values. 
    df.update(species_df)
    
# Show resulting dataframe.
df.head()
    
```

```python
# Search for duplicates.

duplicated = df.duplicated()
df[duplicated]
```

### Apparently the Gentoo species has greater body mass.

```python
from scipy.stats import f_oneway
```

```python
groups = [group['body_mass_g'] for name, group in df.groupby('species')]

# Perform analysis of variance (ANOVA).
statistic, p_value = f_oneway(*groups)

# Show results.
print("Estadístico de prueba:", statistic)
print("Valor p:", p_value)
```

### Since the p-value is extremely small, much less than the significance threshold of 0.05, we can conclude that there is significant evidence to reject the null hypothesis. In other words, at least one group has a mean that is significantly different from the other groups.
### In this context, it is important to consider that even if the p-value is very small, it does not provide information about which specific group or groups have different means, it only indicates that at least one of them does. To resolve this, proceed as follows.

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
# Fit the ANOVA model.
model = ols('body_mass_g ~ species', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

# Perform Tukey post hoc analysis.
tukey_result = sm.stats.multicomp.pairwise_tukeyhsd(df['body_mass_g'], df['species'])

# Display the result of the post hoc analysis.
print(tukey_result)
```

### In summary, the Tukey HSD analysis indicates that there are significant differences in the means between the Adelie and Gentoo groups, as well as between the Chinstrap and Gentoo groups. However, there is no significant difference between the Adelie and Chinstrap groups.
